#Import les bibliotheques :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#Lire les donnees de fichier :

```python
data = pd.read_csv('/content/P4-Movie-Ratings.csv')
data.head()
```

#Les dimension de l'ensemble de donnees :

```python
data.shape
```

#Changement des nom des variables :

```python
data.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']
```

#Informations sur l'ensemble de donnees :

```python
data.info()
```

#Changement des types :

```python
data.Film = data.Film.astype('category')
data.Genre = data.Genre.astype('category')
data.Year = data.Year.astype('category')
data.info()
```

#Les statistiques :

```python
data.describe().T
```

#jointplot de AudienceRating en fonction de CriticRating :

```python
sns.jointplot(data=data, x='CriticRating', y='AudienceRating')
plt.show()
```

```python
sns.jointplot(data=data, x='CriticRating', y='AudienceRating',kind='hex')
plt.show()
```

#Distribution de AudienceRating avec distplot:

```python
m1 = sns.distplot(data.AudienceRating, bins = 15)
```

#Distribution de CriticRating avec distplot:

```python
m2 = sns.distplot(data.CriticRating, bins = 30)
```

#Distribution de AudienceRating avec hist:

```python
sns.set_style('whitegrid') 
n1 = plt.hist(data.AudienceRating, bins=15)
```

#Distribution de Budget:

```python
Filtre1 = data.Genre=='Drama'
Filtre2=data.Genre=='Action'
Filtre3=data.Genre=='Thriller'
plt.hist(data[Filtre2].BudgetMillions, bins = 15,color="red")
plt.hist(data[Filtre1].BudgetMillions, bins = 15,color="green")
plt.hist(data[Filtre3].BudgetMillions, bins = 15,color="blue")
plt.show()
```

```python
gList = data.Genre.cat.categories
vis1 = sns.lmplot(data=data, x = 'CriticRating', y='AudienceRating', fit_reg=False, hue = 'Genre', size = 7, aspect = 1)
plt.show()
```

```python
k1 = sns.kdeplot(data.CriticRating, data.AudienceRating)
plt.show()
```
